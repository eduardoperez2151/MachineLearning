# Forest Fires
---

### Introducción

El objetivo de este problema es determinar en base a 12 atributos relacionados con el clima, que cantidad de hectareas se queman en caso de incendio. La naturaleza del problema es de regresión.
Para resolver este problema utilizaremos las tecnicas de ensamblados, nos basaremos en los modelos de :
- Regresión Lineal
- SVM
- Arboles de regresión
- Random Forest

### Atributos
---

| Atributo |                  Descripción                 | Tipo de Dato |
|:--------:|:--------------------------------------------:|:------------:|
| X        | Coordenada x en el mapa del parque Montesino | Númerico     |
| Y        | Coordenada y en el mapa del parque Montesino | Númerico     |
| month    | Mes del año                                  | Categórico   |
| day      | Día de la semana                             | Categórico   |
| FFMC     | Código de Húmedad de Combustibles Finos      | Númerico     |
| DMC      | Código de Húmedad Duff                       | Númerico     |
| DC       | Código de sequía                             | Númerico     |
| ISI      | Índice de propagación inicial                | Númerico     |
| temp     | Temperatura en grados Celcius                | Númerico     |
| RH       | Húmedad Relativa (%)                         | Númerico     |
| wind     | Velocidad del viento (Km/h)                  | Númerico     |
| rain     | densidad de lluvia (mm/m2)                   | Númerico     |
| area     | área quemada (ha)                            | Númerico     |

```python
import numpy as np
import pandas as pd


#Cargamos el dataset
ds=pd.read_csv("../data/forestfires.csv")

#Revisamos lo tipos de datos del dataset
ds.dtypes
```

```python
#Obtenemos las medidas estaísticas
ds.describe()
```

# Missing Values
---

```python
#Vamos a contar si tenemos atributos faltantes
ds.isna().sum()
```

```python
#A efectos del problema vamos a eliminar las coordendas en el mapa.
ds=ds.drop(["X","Y"],axis=1)

#Chequeamos que se haya realizado correctamente
ds.head()
```

```python
#Como podemos ver en nuestro dataset debemos transformar los meses y dias de String a Númerico
print("meses: %s" % ds.month.unique())
print("días: %s" % ds.day.unique())

#Defino el mapa para la transformación de los meses
month_number= {'jan':1, 'feb':2, 'mar':3, 'apr':4, 'may':5 , 'jun':6, 'jul':7 , 'aug':8, 'sep':9, 'oct':10, 'nov':11, 'dec':12}

#y para el de los días
day_number= {'sun':1, 'mon':2, 'tue':3, 'wed':4, 'thu':5 , 'fri':6, 'sat':7}

ds.month=ds.month.map(month_number)
ds.day=ds.day.map(day_number)

#verifico que se hayan efectuado los cambios
print("\nmeses: %s" % ds.month.unique())
print("días: %s" % ds.day.unique())
```

```python
ds.head()
```

# Dispersión de lo datos
---

```python
import matplotlib.pyplot as plt
import seaborn as sns

def print_dist_plot(dataset):
    for column in dataset.columns:
        sns.distplot(ds[column], kde=True, rug=True)
        plt.show()

print_dist_plot(ds)

ds.area=ds.area.map(np.log)
```

```python
#Creamos una función para imprimir la matriz de correlacion.
def printCorrelationMatrix(dataset):
    correlation=ds.corr()
    mask = np.zeros_like(correlation)
    mask[np.triu_indices_from(mask)] = True
    plt.figure(figsize=(20,20))
    with sns.axes_style("white"):
        ax = sns.heatmap(correlation, mask=mask,cmap="Blues",annot=True)
        
printCorrelationMatrix(ds)
```

```python
#Creamos una función para imprimir lo diagraas de caja
def printBoxplot(dataset):
    for column in dataset.columns:
        print("\n####### "+str(column)+" #######")
        sns.boxplot(dataset[column])
        plt.show()
        
#los imprimimos
printBoxplot(ds.drop(['area'],axis=1))
```

```python
ds=ds.replace([-np.inf], 0)
ds.quantile([.25,.5,.75,.95])

ds.dtypes
```

### Normalización
---

```python
#Importo el Normalizador de SKLearn
from sklearn.preprocessing import Normalizer

#Quito la variable de predicción
X= np.array(ds.drop(['area'],1))

#Ajusto nuestro normalizador a 
#nuestros datos y luego aplicando la transformación
transformer = Normalizer().fit(X)
X=transformer.transform(X)

#Verificamos la Normalización
print(X)
```

### Entrenamiento y Evaluación de los modelos

```python
from sklearn.linear_model import LinearRegression
from sklearn.svm import SVR
from sklearn.tree import DecisionTreeRegressor
from sklearn.ensemble import RandomForestRegressor
from sklearn import metrics

from sklearn import model_selection

#Me quedo con la variable de prediccion
Y = ds['area']

#Ajusto el tamaño de los datos para test
validation_size = 0.30
#Utilizo una semilla para poder reproducir el split
seed = 2018

#Hago un split para validar el modelo
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X,Y, test_size=validation_size, random_state=seed)

def predict(model,x_train,y_train,x_validation,y_validation,name):

    #Ajusto el modelo a mis datos
    model.fit(x_train,y_train)

    predictions=model.predict(x_validation)
    print("%s:" % name)
    print("MSE : %s\n" % np.sqrt(metrics.mean_squared_error(y_validation, predictions)))
    
#Definó el modelo de Árbol de desición
model_tree = DecisionTreeRegressor()
predict(model_tree,X_train,Y_train,X_validation,Y_validation,"Árbol de Regresión")

#Definó el modelo de regresión Líneal
model_lineal=LinearRegression()
predict(model_lineal,X_train,Y_train,X_validation,Y_validation,"Regresión Líneal")

#Definó el modelo de SVM
model_lineal=SVR(gamma='auto')
predict(model_lineal,X_train,Y_train,X_validation,Y_validation,"Maquina de Soporte Vectorial")
```
