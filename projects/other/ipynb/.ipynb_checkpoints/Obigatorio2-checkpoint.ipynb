### Atributos Faltantes
--- 
En el dataset los atributos faltantes se denotan con el caracter **"?"** vamos a buscar dichos valores y en base a los mismos **determinar que estrategia tomar para tratarlos**.

```python
#Reemplazamos todos los caracteres "?" por Numpy.NaN
ds=ds.replace("?",np.nan)

'''
Contamos los valores faltantes y obtenemos un mapa con el nombre 
del atributo y la cantidad de valores faltantes
'''
def count_missing_values(dataset,total_instances=1):
    for column in dataset.columns:
        count=ds[column].isna().sum()
        if count >0:
            print ("%s: %s" % (column, str(round(count/total_instances, 4))))

#Cotamos los missing values
print("\n#### Cantidad de atributos faltantes ####\n")
count_missing_values(ds)

print("\n#### porcentaje de atributos faltantes ####\n")
#Vamos a obtener el porcentaje de los dataos faltates
count_missing_values(ds,total_instances);
```

Como podemos observar es imposible recontruir la informacion del peso **(weight)** al igual que el de la especialidad del Medico que admite al paciente **(medical_specialty)** , debido a la enorme cantidad de dato faltantes imputar valores seria un gran error ya que podriamos caer en el error de sesgar nuestro dataset.
Por lo que vamos a eliminiar estos atributos.

```python
#Eliminamos los dos atributos
ds=ds.drop(['weight','medical_specialty'],axis=1)

#nuevamente vamos a chequear nustros valores faltantes
count_missing_values(ds)
```

Dichos atributos tienen la caracteristica que son nominales, por lo que vamos a imputar valores utilizando las estrategia de los **mas frecuentes** que nos provee **skLearn**

```python

from sklearn_pandas import CategoricalImputer


#Obtengo los headers de mi dataset
headers=ds.columns


#Creo el objeto encargado de hacer las transformacion de mis datos (imputar los datos)
imputer = CategoricalImputer()

#Aplico la transformacion(imputacion de datos) y devulvo un nuevo DataFrae
ds.race=imputer.fit_transform(ds.race)
ds.diag_1=imputer.fit_transform(ds.diag_1)
ds.diag_2=imputer.fit_transform(ds.diag_2)
ds.diag_3=imputer.fit_transform(ds.diag_3)

#Chequeamos valores faltantes
ds.isna().sum()
```

Como podemos observar no tenemos valores faltantes, Ahora veremos las distribuciones del dataset

### Normalizacion
---
Como tenemos atributos categoricos y númericos en nuestro dataset la Normalización unicamente se la vamos a aplicar para aquelos atributos númericos.

```python
#Importo el Normalizador de SKLearn
from sklearn.preprocessing import Normalizer



def normalize(dataset,columns):
  for column in columns:
    #Ajusto nuestro normalizador a 
    #nuestros datos y luego aplicamos la transformación todo en un mismo paso
    dataset[column] = Normalizer().fit_transform(dataset[columns])

    
#Realizamos la Normalización
normalize(ds,columns)

#imprimo los primeros 20 elementos del dataset para verificar la normalización
ds.head(20)
```

### Discretización de los datos
---
Vamos a realizar la discretizacion de los datos para ello utilizaremo KBinDiscretizer ,lo dejaremos con 5 bins por defecto, pero debemos ser concientes que esto puede afectarnos a la hora de acer la prediccion, el Bining o discretización se debe de hacer con mucho cuidado revisando las distribuciones de los valores de cada atributo y en base a ellas crear el el bucket (rango).

```python
def discretize(dataset,columns):
  for column in columns:
    #Ajusto nuestro normalizador a 
    #nuestros datos y luego aplicamos la transformación todo en un mismo paso
    dataset[column] =pd.cut(dataset[column], bins=5)
    
    
#Realizamos la Normalización
discretize(ds,columns)
```

## Modelado
---
Para la etapa de modelado vamos a utilizar la estrategia de ensables de clasificadores, los modelos que vamos a utilizar son Arboles de desiciones, Random forest y  SVM. Comencemos!!!




```python
from sklearn import tree
from sklearn import model_selection
from sklearn import preprocessing

label_encoder = preprocessing.LabelEncoder()
ds= ds.apply(label_encoder.fit_transform)

#Me quedo con la variable de prediccion
Y = np.array(ds['readmitted'])

#Ajusto el tamaño de los datos para test
validation_size = 0.30

#Utilizo una semilla para poder reproducir el split
seed = 2018

X=np.array(ds.drop(['readmitted'],axis=1))


#Hago un split para validar el modelo
X_train, X_validation, Y_train, Y_validation = model_selection.train_test_split(X,Y, test_size=validation_size, random_state=seed)

#Defino el KFold
kfold = model_selection.KFold(n_splits=10, random_state=seed)

#Defino el modelo
model_dtree = tree.DecisionTreeClassifier(criterion = "gini",max_depth=15)
#Obtengo los resultados del entrenamiento 
results_dtree = model_selection.cross_val_score(model_dtree, X_train,Y_train, cv=kfold, scoring='accuracy')
print("#### Resultado Training ####\n")
print("Media: %f" % (results_dtree.mean()))
print("Desvio Estandar: %f" % (results_dtree.std()))
```

```python
from sklearn.ensemble import RandomForestClassifier

#Defino el modelo de Random fores con un boque de 400 arboles ada uno con una profundidad de 7 nodos
model_rnd_forest=RandomForestClassifier(n_estimators=400,  max_depth=7)

#Obtengo los resultados del entrenamiento 
results_rnd_forest = model_selection.cross_val_score(model_rnd_forest, X_train,Y_train, cv=kfold, scoring='accuracy')
print("#### Resultado Training ####\n")
print("Media: %f" % (results_rnd_forest.mean()))
print("Desvio Estandar: %f" % (results_rnd_forest.std()))
```

```python
from sklearn import svm

#Defino un modelo de SVM
model_svm=svm.LinearSVC()

#Obtengo los resultados del entrenamiento 
results_svm = model_selection.cross_val_score(model_rnd_forest, X_train,Y_train, cv=kfold, scoring='accuracy')
print("#### Resultado Training ####\n")
print("Media: %f" % (results_svm.mean()))
print("Desvio Estandar: %f" % (results_svm.std()))
```

```python
#Ahora vamos a crear nuestro ensable utilizando los tres modelos antes definios definidos
 from sklearn.ensemble import VotingClassifier
    
model_ensemble=VotingClassifier(estimators=[('dt', model_dtree), ('rf', model_rnd_forest), ('sv', model_svm)], voting='hard')

#Obtengo los resultados del entrenamiento 
resultsmodel_ensemble = model_selection.cross_val_score(model_ensemble, X_train,Y_train, cv=kfold, scoring='accuracy')
print("#### Resultado Training ####\n")
print("Media: %f" % (results_svm.mean()))
print("Desvio Estandar: %f" % (results_svm.std()))
```

## Evaluación
---
Una vez que creamos nuestros modelos y nuestro ensamble vamos a evaluarlos con datos que nuestros modelo nunca han visto.

```python
#predecimos con nuestro arbol de desición
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


#Lo ajusto a los datos de entrenamiento
model_dtree.fit(X_train, Y_train)

#Hago las preciiones sbre los datos de Validacion
predictions_dtree = model_dtree.predict(X_validation)

#Agregó el scoring a un array para luego hacer una gráfica
accuracy_dtree=accuracy_score(Y_validation, predictions_dtree)

#imprimimos el Accurcy
print("\nAccuracy: %s\n" % accuracy_dtree)

#imprimimos otras metricas del model
print(classification_report(Y_validation, predictions_dtree))
```

```python
#predecimos con nuestro random forest
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


#Lo ajusto a los datos de entrenamiento
model_rnd_forest.fit(X_train, Y_train)

#Hago las preciiones sbre los datos de Validacion
predictions_rnd_forest = model_rnd_forest.predict(X_validation)

#Agregó el scoring a un array para luego hacer una gráfica
accuracy_rnd_forest=accuracy_score(Y_validation, predictions_rnd_forest)

#imprimimos el Accurcy
print("\nAccuracy: %s\n" % accuracy_rnd_forest)

#imprimimos otras metricas del model
print(classification_report(Y_validation, predictions_rnd_forest))
```

```python
#predecimos con nuestro SVM
from sklearn.metrics import accuracy_score
from sklearn.metrics import classification_report


#Lo ajusto a los datos de entrenamiento
model_svm.fit(X_train, Y_train)

#Hago las preciiones sbre los datos de Validacion
predictions_svm = model_svm.predict(X_validation)

#Agregó el scoring a un array para luego hacer una gráfica
accuracy_svm=accuracy_score(Y_validation, predictions_svm)

#imprimimos el Accurcy
print("\nAccuracy: %s\n" % accuracy_svm)

#imprimimos otras metricas del model
print(classification_report(Y_validation, predictions_svm))
```
